```python
import requests
import json
r =requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text

response
```

這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

```python
data=json.loads(response)

print(len(data))

print(data[0].keys())
```

取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

```python
import json
data=json.loads(response)

for d in data:
    print('標題: ',d['title'])
    print('貼文時間: ',d['createdAt'])
    print('留言人數: ',d['commentCount'])
    print('按讚人數: ',d['likeCount'])
    
```

計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
import requests
import json
r=requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
res=r.text
res=json.loads(res)
comment=0
like=0
for n in res:
    comment+=n['commentCount']
    like=like+n['likeCount']
print('熱門文章平均留言人數:',comment/len(res),'平均按讚人數:',like/len(res))    

r=requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
res=r.text
res=json.loads(res)
comment=0
like=0
for n in res:
    comment=comment+n['commentCount']
    like=like+n['likeCount']
print('非熱門文章平均留言人數:',comment/len(res),'平均按讚人數:',like/len(res))    
```
